```python
import gdsfactory as gf
```

```python
grating = gf.components.grating_coupler_elliptical(polarization='te').copy()

# grating.draw_ports()
grating.pprint_ports()

grating.plot()
```

```python
mzi_base = gf.components.mzi(bend='bend_euler', straight='straight', length_y=10, splitter='mmi1x2', with_splitter=True, port_e1_splitter='o2', port_e0_splitter='o3', port_e1_combiner='o2', port_e0_combiner='o3', port1='o1', port2='o2', nbends=2, cross_section='strip', mirror_bot=False, add_optical_ports_arms=False, min_length=0.01, auto_rename_ports=True).copy()

# mzi_base.draw_ports()
mzi_base.pprint_ports()

mzi_base.plot()
```

```python
ring_mod = gf.components.ring_single_pn(gap=0.4, radius=10, doping_angle=250, doped_heater=True, doped_heater_angle_buffer=10, doped_heater_layer='NPP', doped_heater_width=0.5, doped_heater_waveguide_offset=1.175).copy()

wg = gf.components.straight(length=30, cross_section='strip').copy()

pn_ring = gf.Component()

pad = gf.components.pad(size=(5, 5), layer=44).copy()
pad_ring = gf.components.pad(size=(1, 1), layer=44).copy()

ring = pn_ring << ring_mod
padring1 = pn_ring << pad_ring
padring2 = pn_ring << pad_ring
wg1 = pn_ring << wg
wg2 = pn_ring << wg
pad1 = pn_ring << pad
pad2 = pn_ring << pad

wg1.connect("o2", ring.ports["o1"])
wg2.connect("o1", ring.ports["o2"])

padring1.x = pn_ring.x - 10
padring2.x = pn_ring.x - 6.3
padring1.y = pn_ring.y - 6.5
padring2.y = pn_ring.y - 4

pad1.x = pn_ring.x - 10
pad2.x = pn_ring.x + 10
pad1.y = pn_ring.y - 25
pad2.y = pad1.y

pn_ring.add_port("o1", wg1.ports["o1"])
pn_ring.add_port("o2", wg2.ports["o2"])

pn_ring.add_port("e1_heater", padring1.ports["e4"])
pn_ring.add_port("e2_heater", padring2.ports["e4"])
pn_ring.add_port("A", pad1.ports["e2"])
pn_ring.add_port("GND", pad2.ports["e2"])

gf.routing.route_single_electrical(component=pn_ring, port1=pad1.ports["e2"], port2=padring1.ports["e4"], start_straight_length=3, end_straight_length=3, width=1, cross_section='metal1')

gf.routing.route_single_electrical(component=pn_ring, port1=pad2.ports["e2"], port2=padring2.ports["e4"], start_straight_length=3, end_straight_length=3, width=1, cross_section='metal1')

# pn_ring.draw_ports()
pn_ring.pprint_ports()

pn_ring.plot()
```

```python
c = gf.Component()

mzi = c << mzi_base
gcIn = c << grating
gcOut = c << grating
ring1 = c << pn_ring
ring2 = c << pn_ring

gcIn.connect("o1", mzi.ports["o1"])
gcOut.connect("o1", mzi.ports["o2"])

ring1.rotate(180)
ring1.x = mzi.x
ring2.x = mzi.x
ring1.ymin = mzi.ymax + 5
ring2.ymax = mzi.ymin - 5

c.add_port("te_gcIn", gcIn.ports["o2"])
c.add_port("te_gcOut", gcOut.ports["o2"])
c.add_port("o1_ring1", ring1.ports["o2"])
c.add_port("o2_ring1", ring1.ports["o1"])
c.add_port("o1_ring2", ring2.ports["o1"])
c.add_port("o2_ring2", ring2.ports["o2"])
c.add_port("A1", ring1.ports["A"])
c.add_port("GND1", ring1.ports["GND"])
c.add_port("A2", ring2.ports["A"])
c.add_port("GND2", ring2.ports["GND"])

c.draw_ports()
c.pprint_ports()

c.plot()
```
